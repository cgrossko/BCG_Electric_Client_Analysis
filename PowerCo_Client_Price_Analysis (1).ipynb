```python
client_df['date_activ'] = pd.to_datetime(client_df['date_activ'])
```

```python
# Bar plot of churn counts
plt.figure(figsize=(6, 4))
sns.countplot(data=client_df, x='churn')
plt.title('Churn Counts')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()

# Pie chart of churn proportion
churn_counts = client_df['churn'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Churn Proportion')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

```

```python
# Churn rate over time (assuming date is available)
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=client_df, x='date_activ', hue='churn')#, estimator='mean
plt.title('Churn Rate Over Time')
plt.xlabel('Activation Date')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)
ax.xaxis.set_major_locator(plt.MaxNLocator(10))  # Adjust the tick frequency
plt.tight_layout()
plt.show()

# Bar plot of churn by 'channel_sales'
plt.figure(figsize=(15, 10))
sns.countplot(data=client_df, x='channel_sales', hue='churn')
plt.title('Churn by Channel Sales')
plt.xlabel('Channel Sales')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Churn')
plt.show()
```

```python
# Box plot of churn by 'cons_12m'
plt.figure(figsize=(10, 6))
sns.boxplot(data=client_df, x='churn', y='cons_12m')
plt.title('Churn vs. cons_12m')
plt.xlabel('Churn')
plt.ylabel('cons_12m')
plt.ylim(-10000, 100000)
plt.show()
```
